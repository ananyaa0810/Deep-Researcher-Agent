```python
import os
import nltk
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
from nltk.tokenize import sent_tokenize

# Update this path after uploading your PDF
PDF_PATH = "/content/1.pdf"  # Replace with your uploaded file name

def read_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

pdf_text = read_pdf(PDF_PATH)
print("PDF loaded successfully! Total characters:", len(pdf_text))
```

```python
def chunk_text(text, chunk_size=500):
    sentences = sent_tokenize(text)
    chunks, current_chunk = [], ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) <= chunk_size:
            current_chunk += sentence + " "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + " "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

chunks = chunk_text(pdf_text)
documents = [
    {'id': f'doc_{i}', 'text': chunk, 'meta': {'source': os.path.basename(PDF_PATH), 'chunk_idx': i}}
    for i, chunk in enumerate(chunks)
]

print(f"Total chunks created: {len(chunks)}")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')  # Small, efficient embedding model
embeddings = model.encode([doc['text'] for doc in documents])
print("Embeddings generated successfully!")
```

```python
embedding_dim = embeddings.shape[1]
index = faiss.IndexFlatL2(embedding_dim)
index.add(np.array(embeddings))
print(f"FAISS index created with {index.ntotal} documents.")
```

```python
def multi_step_query(query, steps=2, top_k=3):
    step_results = []
    sub_queries = query.split(".")[:steps]  # Naive breakdown for now
    for step in sub_queries:
        if step.strip():
            query_vector = model.encode([step])
            distances, indices = index.search(query_vector, top_k)
            for idx, score in zip(indices[0], distances[0]):
                step_results.append({
                    'sub_query': step,
                    'text': documents[idx]['text'],
                    'score': float(score)
                })
    return step_results

query = "Summarize the introduction. Explain key challenges mentioned."
results = multi_step_query(query)
for res in results:
    print(f"\nSub-query: {res['sub_query']}\nScore: {res['score']:.4f}\nText: {res['text']}")
```

```python
from sentence_transformers import util

def summarize_results(results, max_sentences=5):
    combined_text = " ".join([res['text'] for res in results])
    sentences = sent_tokenize(combined_text)
    return " ".join(sentences[:max_sentences])

summary = summarize_results(results)
print("\nGenerated Summary:\n", summary)
```

```python
import ipywidgets as widgets
from IPython.display import display

def interactive_query_widget():
    query_box = widgets.Text(
        value='',
        placeholder='Type your query here',
        description='Query:',
        disabled=False
    )

    output_area = widgets.Output()

    def on_submit(change):
        with output_area:
            output_area.clear_output()
            results = multi_step_query(change['new'])
            summary = summarize_results(results)
            print("\nSummary of results:\n", summary)

    query_box.observe(on_submit, names='value')
    display(query_box, output_area)

# Run this to get a nice interactive interface
interactive_query_widget()

```

```python
def explain_reasoning(query):
    steps = query.split(".")
    explanation = "Reasoning steps:\n"
    for i, step in enumerate(steps, 1):
        explanation += f"Step {i}: '{step.strip()}' -> Search for relevant chunks.\n"
    return explanation

reasoning = explain_reasoning(query)
print(reasoning)
```
